```
import pandas as pd
```

```
파일경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris2.csv'
아이리스 = pd.read_csv(파일경로)
아이리스.head()
```

```
print(아이리스.dtypes)
```

```
# 품종 타입을 범주형으로 바꾸기
아이리스['품종'] = 아이리스['품종'].astype('category')
print(아이리스.dtypes)
```

```
# NA값 체크
아이리스.isna().sum()
```

```
아이리스.tail()
```

```
# NA값에 평균값을 넣어주기. (가장 좋은 방법은 NA값이 많지 않을때 엑셀을 통해서 직접 값을 넣어주는 것이지만 임의로 평균값을 넣어줄 수 있음)
mean = 아이리스['꽃잎폭'].mean()
mean
아이리스['꽃잎폭'] = 아이리스['꽃잎폭'].fillna(mean)
아이리스.tail()
```
